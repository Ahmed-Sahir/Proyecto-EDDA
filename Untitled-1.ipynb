## üéØ Objetivo

Analizar si existe desigualdad regional en el acceso a los servicios de salud en Espa√±a, usando datos de renta, ruralidad, envejecimiento poblacional y n√∫mero de m√©dicos disponibles.

## üß† Hip√≥tesis

1. Las comunidades con menor renta per c√°pita tienen menos m√©dicos por cada 1.000 habitantes.
2. Las regiones con mayor envejecimiento tienen m√°s necesidad m√©dica.
3. Las zonas rurales presentan menos acceso a servicios sanitarios.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"]= (10, 6)
```

```python
renta = pd.read_csv('/Users/user/Downloads/renta_ccaa.csv')
medicos = pd.read_csv('/Users/user/Downloads/medicos_ccaa.csv')
envejecimiento = pd.read_csv('/Users/user/Downloads/envejecimiento_ccaa.csv')
ruralidad = pd.read_csv('/Users/user/Downloads/ruralidad_ccaa.csv')
```

```python
renta.head()
medicos.describe()
envejecimiento.isnull().sum()
ruralidad.info()
```

```python
df = renta.merge(medicos, on = "CCAA")
df = df.merge(envejecimiento , on = "CCAA")
df = df.merge(ruralidad, on = "CCAA")

df.head()
```

```python
print(df.columns)
```

```python
import os

os.makedirs("figures", exist_ok=True)
sns.scatterplot(data=df, x='RentaMedia', y='MedicosPorMil', hue='CCAA')
plt.title('Renta vs M√©dicos por 1000 habitantes por CCAA')
plt.savefig("figures/renta_vs_medicos.png")

plt.show()
```

```python
sns.regplot(data=df, x="PorcentajeMayores65", y="MedicosPorMil", ci=None)
plt.title("Envejecimiento vs M√©dicos por 1.000 habitantes")
plt.savefig("figures/envejecimiento_vs_medicos.png")
plt.show()
```

```python
sns.scatterplot(data=df, x="PorcentajePoblacionRural", y="MedicosPorMil", hue="CCAA")
plt.title("Ruralidad vs M√©dicos por cada 1.000 habitantes")
plt.savefig("figures/ruralidad_vs_medicos.png")
plt.show()
```

## ‚úÖ Conclusiones

- Existe una relaci√≥n **moderada** entre renta y acceso m√©dico: comunidades con mayor renta suelen tener m√°s m√©dicos.
- Las regiones con poblaci√≥n envejecida **no siempre** tienen mejor cobertura, lo que contradice la hip√≥tesis.
- La **ruralidad influye** claramente: a mayor porcentaje de poblaci√≥n rural, menor acceso a servicios m√©dicos.

Esto sugiere que factores geogr√°ficos y de infraestructura son tan importantes como la renta para explicar la desigualdad sanitaria.
